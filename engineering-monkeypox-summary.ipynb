~ 100 words ~

## Design
The client, the [Centers for Disease Control and Prevention](https://www.cdc.gov/) wants know what people are saying about [monkeypox](https://www.cdc.gov/poxvirus/monkeypox/index.html) on Twitter to better inform their communication campaign.<br/>

**Business Problem:** How to monitor chatter on Twitter about monkeypox?  <br/>

**Pipeline**: To monitor chatter on Twitter, need to built a proof of concept dashboard.<br/>

**Impact Hypothesis:** Inform CDC's communication strategy on monkeypox to avoid stigma.<br/>

## Data
Kaggle [Twitter Dataset on the 2022 MonkeyPox Outbreak](https://www.kaggle.com/datasets/thakurnirmalya/monkeypox2022tweets) Dataset is list of TweetIDs (n = 229,185) which where [hydrated](https://github.com/DocNow/hydrator/releases) to access the full tweet information. This is to comply with Twitter's developer terms to not publicly post private information (i.e., handles).<br/>
