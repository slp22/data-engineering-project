# Monkeypox Tweets

## Question
* What is the end service that your project will provide? What is the purpose of the system you plan to build?
    *  Dashboard to monitor top trending topics on Twitter about monkeypox.
* Who is your client and how will that client benefits from your end service?
    * The client, Centers for Disease Control and Prevention, wants know what people are saying about [monkeypox](https://www.cdc.gov/poxvirus/monkeypox/index.html) on Twitter to better inform their communication campaign. 
* What does your end-to-end data pipeline look like? Submit a preliminary sketch of your pipeline in the form of a diagram or workflow list with your proposal. 

## Data Description
* What dataset(s) do you plan to use, and how will you obtain the data? 
    * Kaggle: [Tweets on monkeypox](https://www.kaggle.com/datasets/aneeshtickoo/tweets-on-monkeypox) (n = 6,859)
    * Filtered for `language` = English (n = 6,410)
* Do you plan to be able to automatically pull in new data at a regular cadence (e.g., Airflow or cron job)?
    * No 
What is an individual sample/unit of analysis in this project?
    * One tweet


* What characteristics/features do you expect to work with?
    * `language`
    * `date`
    * `username`
    * `hashtags`
    * `tweet`

* If modeling, what will you predict as your target?
    * Topic trends across time
    * Frequency of CDC-defined [symptoms](https://www.cdc.gov/poxvirus/monkeypox/symptoms.html) mentions: rash, fever, chills
